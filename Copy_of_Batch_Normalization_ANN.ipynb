```python
import pandas as pd
import numpy as np
```

```python
Data=pd.read_csv("/content/concertriccir2.csv")
```

```python
Data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(Data['X'],Data['Y'],c=Data['class'])
plt.show()
```

```python
X = Data.iloc[:,0:2].values
Y = Data.iloc[:,-1].values
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

```python
# ANN Architecture...
model = Sequential()
model.add(Dense(3,activation='relu',input_dim=2))
model.add(Dense(3,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
#Compling the model...
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Training my model...
history1=model.fit(X,Y,epochs=200,validation_split=0.2)
```

```python
plt.plot(history1.history["accuracy"],color="red")
plt.plot(history1.history["val_accuracy"],color="blue")
plt.show()
```

```python
# ANN Arhitecture...
model = Sequential()
model.add(Dense(2,activation='relu',input_dim=2))
model.add(BatchNormalization())
model.add(Dense(2,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1,activation='sigmoid'))
```

```python
# compliling the model..
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Trainingt the model..
history2=model.fit(X,Y,epochs=200,validation_split=0.2)
```

```python
plt.plot(history2.history["accuracy"],color="red")
plt.plot(history2.history["val_accuracy"],color="black")
plt.show()
```

```python
# Accuracy as per the Batch Normalizarion and without Normalization...
plt.plot(history1.history["accuracy"],color="red")  # Accuracy without Normalization..
plt.plot(history2.history["accuracy"],color="black")  # Accuracy with Normalization..
plt.show()
```

```python
# Validation Accuracy as per the Batch Normalizarion and without Normalization...
plt.plot(history1.history["val_accuracy"],color="red")  # Val Accuracy without Normalization..
plt.plot(history2.history["val_accuracy"],color="black")  # Val Accuracy with Normalization..
plt.show()
```
