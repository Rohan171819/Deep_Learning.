```python
import pandas as pd
import numpy as np
```

```python
Data=pd.DataFrame([[8,8,3],[6,9,5],[7,5,8],[12,7,11]],columns=["iq","cgpa","lpa"])
Data
```

```python
def initialize_parameters(layer_dims):
  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)
  for l in range(1, L):
    parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))
  return parameters
```

```python
initialize_parameters([2,3,1])
```

```python
def linear_forward(A_prev, W, b):
  Z = np.dot(W.T, A_prev) + b
  return Z
```

```python
# Forward Propagation...
def L_layer_forward(X, parameters):

  A = X
  L = len(parameters) // 2                  # number of layers in the neural network

  for l in range(1, L+1):
    A_prev = A
    Wl = parameters['W' + str(l)]
    bl = parameters['b' + str(l)]
    print("A"+str(l-1)+": ", A_prev)
    print("W"+str(l)+": ", Wl)
    print("b"+str(l)+": ", bl)
    print("--"*20)

    A = linear_forward(A_prev, Wl, bl)
    print("A"+str(l)+": ", A)
    print("**"*20)

  return A,A_prev
```

```python
X=Data[["iq","cgpa"]].values[0].reshape(2,1)
y=Data[["lpa"]].values[0][0]

# Parameters Initialization...
parameters=initialize_parameters([2,3,1])
Y_hat,A1=L_layer_forward(X,parameters)
```

```python
Y_hat=Y_hat[0][0]
```

```python
def update_parameters(parameters,y,Y_hat,A1,X):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 *(y-Y_hat) * A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 *(y-Y_hat) * A1[1][0])
  parameters['W2'][2][0] = parameters['W2'][2][0] + (0.001 * 2 *(y-Y_hat) * A1[2][0])
  parameters['b2'][0][0] = parameters['b2'][0][0] + (0.001 * 2 *(y-Y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][0][0] * X[0][0]) #iq
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][1][0] * X[0][0]) #iq
  parameters['W1'][0][2] = parameters['W1'][0][2] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][2][0] * X[0][0]) #iq
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][0][0])           #iq

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][0][0] * X[1][0]) #cgpa
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][1][0] * X[1][0]) #cgpa
  parameters['W1'][1][2] = parameters['W1'][1][2] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][2][0] * X[1][0]) #cgpa
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][1][0]) #cgpa
  parameters['b1'][2][0] = parameters['b1'][2][0] + (0.001 * 2 *(y-Y_hat) * parameters['W2'][2][0]) #cgpa
```

```python
update_parameters(parameters,y,Y_hat,A1,X)
```

```python
# epochs implementation

parameters = initialize_parameters([2,3,1])
epochs = 10

for i in range(epochs):

  Loss = []

  for j in range(Data.shape[0]):

    X = Data[['iq', 'cgpa']].values[j].reshape(2,1) # Shape(no of features, no. of training example)
    y = Data[['lpa']].values[j][0]

    # Parameter initialization
    y_hat,A1 = L_layer_forward(X,parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters,y,y_hat,A1,X)

    Loss.append((y-y_hat)**2)

  print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())

parameters
```
